```python
import numpy as np
import os
```

```python
def is_stochastic(data):
    l = len(data)
#     print(data)
    temp = np.ones(shape= (1,l))
    temp = temp/l
#     print(temp)
    var = 'The given adjacent matrix is already Stochastic'
    
    for i in range(l):
        if np.any(data[i]) == False:
            data[i] = temp
            var = 'The given adjacent matrix is not Stochastic'

    print(var)
    return data
            
```

```python
def print_format(data,result_file):
    i = 0
    l = len(data)
    temp ='('
#     result_file.write(temp)
#     result_file.write('\n')
    while i < l:
        for j in range(l):
            if i == l-1 and j == l-1:
                temp2  = str(round(data[i][j],2)) + ')'
                temp += temp2
            else:
                temp2  = str(round(data[i][j],2)) + ' '
                temp += temp2
        print(temp)
        result_file.write(temp)
        result_file.write('\n')
        temp =''
        i += 1
        
```

```python
def irr_ape(data, d):
    l = len(data)
    E = np.ones(shape = (l,l))
#     print(E)
    temp = ((1-d)*(E/l)) + (d * np.transpose(data))
#     print(temp)
    return temp
    
```

```python
def page_rank(data, k):
    l = len(data)
    e = np.ones(shape = (l,1))
#     print(e)
    p = e/l
    i = 0
#     print(p)
    while i < k:
        p = np.matmul(data, p)
        i += 1
    return p
    
```

```python
def page_rank2(data, d, k):
    l = len(data)
    e = np.ones(shape = (l,1))
#     print(e)
    p = e/l
    i = 0
#     print(p)
    while i < k:
        p = ((1-d)*e) + (d*np.matmul(np.transpose(data), p))
        i += 1
    return p
    
```

```python
def print_format2(data,result_file):
    data = data.flatten()
    temp = '('
#     result_file.write(temp)
#     result_file.write('\n')
    l = len(data)
    for i in range(l):
        if i == len(data)-1:
            temp += (str(round(data[i],2)) + ')')
        else:
            temp += (str(round(data[i],2)) + ' ')

    print(temp)
    result_file.write(temp)
```

```python
def pg(s_id, file_path, result_path,sampletxt,d, k):
    # Loading the data from the given text file

    temp = np.loadtxt(fname = os.path.join(file_path, sampletxt), delimiter= '\n', dtype= str, encoding = 'utf-8-sig')
    data_temp = [i.replace('\t',' ').replace('(','').replace(')', '').split(' ') for i in temp]
    # print(data_temp)
    data = np.asarray([[float(i) for i in j] for j in data_temp])
    

    result_file = open(result_path, "w")
#     result_file.write("(")
    
    print(s_id)
    print('\n')
    result_file.write(s_id)
    result_file.write('\n')
    
    
    print('Q A')
    print("The given adjacent matrix is: ")
#     result_file.write('Q A\n')
#     result_file.write("The given adjacent matrix is:\n")

    for i in range(len(data)):
        if np.sum(data[i]) > 1:
            data[i] = data[i]/np.sum(data[i])
    print_format(data,result_file)
    print('\n')
#     result_file.write('\n')


    print('Q B')     
    data_st = is_stochastic(data)
    print("The stochastic matrix is: ")
#     result_file.write('Q B\n')
#     result_file.write("The stochastic matrix is:\n")
    print_format(data_st,result_file)
    print('\n')
#     result_file.write('\n')

    
    print('Q C')     
    f_data = irr_ape(data_st, d)
    print("The final matrix is: ")
#     result_file.write('Q C\n')
#     result_file.write("The final matrix is: \n")
    print_format(f_data,result_file)
    print('\n')
#     result_file.write('\n')

    
    print('Q D')
    print("The page rank vector is: ")
    output = page_rank(f_data, k)
#     result_file.write('Q D\n')
#     result_file.write("The page rank vector is: \n")
    print_format2(output,result_file)
    
    
    
```
